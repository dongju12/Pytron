# 문제4

```python
score_list = [['김미희', 89,78,99],
             ['서대희', 67,88,77],
             ['강예진',100,55,96],
             ['차현석',75,100,60]]
                
```

```python
score_list
```

```python
score = pd.DataFrame(score_list, columns = ['이름', 'Python', 'Java', 'DB'])
score
```

```python
score['평균']=score[['Python','Java','DB']].mean(axis = 1)
score
```

```python
def makeGd(row):
    mean = row['평균']
    
    if mean >= 90 :
        return 'A'
    elif mean >= 80:
        return 'B'
    elif mean >= 70:
        return 'C'
    elif mean >= 60:
        return 'D'
    else:
        return 'F'
    
score['학점'] = score.apply(makeGd, axis = 1)
score
```

# 문제6

```python
data = pd.read_csv('./data/Traffic_Accident_2017.csv', encoding = 'euc-kr')
data.head()
```

```python
ctp = data[data['사고유형_대분류'] == '차대사람']
ctp
```

```python
time = ctp[ctp['주야'] == '야간']
```

```python
temp = time['요일'].value_counts()
y = temp[['월','화','수','목','금','토','일']]
y
```

```python
x=y.index
x
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.bar(x,y)
plt.title("요일별 야간 차대사람 교통사고 건수")
plt.xlabel("요일")
plt.ylabel("사고 건수")
plt.ylim(0,200)

plt.show()
```
